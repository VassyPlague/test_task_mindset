```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_path = 'DataForOCR/0.jpeg'
#img_path = '1.jpg'
if img_path == 'DataForOCR/8.jpeg':
    no_cut_img = cv2.imread(img_path)
    img = no_cut_img[int (no_cut_img.shape[0]/2) : no_cut_img.shape[0], 0 : int (no_cut_img.shape[1]/2) ]
else:
    no_cut_img = cv2.imread(img_path)
    img = no_cut_img[int (no_cut_img.shape[0]/2):no_cut_img.shape[0], 0 : no_cut_img.shape[1] ]
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, edge = cv2.threshold(img_gray, 90, 255, cv2.THRESH_BINARY)
plt.imshow(edge)
```

```python
# verctical
h = int (edge.shape[0]/15)
w = int (edge.shape[1]/2 + edge.shape[1]/6)
vert_border = -1 
while w < edge.shape[1]:
    while h < edge.shape[0]:
        if edge[h][w] == 255 :
            if h < edge.shape[0]/2 - edge.shape[0]/3 :
                vert_border = h
                break
        h += 1
    w += 1

print ('vert = ' +  str (vert_border) + ' wide = ' + str(w))    

if vert_border == -1:
    h = int (edge.shape[0]/15)
    w = int (edge.shape[1]/2)
    while w < edge.shape[1]:
        while h < edge.shape[0]:
            if edge[h][w] == 255:
                if h < edge.shape[0]/2 - edge.shape[0]/3:
                    vert_border = h
                    break
            h += 1
        if vert_border != -1:
                break
        w += 1
        
print ('vert = ' +  str (vert_border) + ' wide = ' + str(w)) 
        
if vert_border == -1:
    h = int (edge.shape[0]/20)
    w = int (edge.shape[1]/2)
    while w < edge.shape[1]:
        while h < edge.shape[0]:
            if edge[h][w] == 255:
                if h < edge.shape[0]/2 - edge.shape[0]/3:
                    vert_border = h
                    break
            h += 1
        if vert_border != -1:
            break
        w += 1        

print ('vert = ' +  str (vert_border) + ' wide = ' + str(w))
```
