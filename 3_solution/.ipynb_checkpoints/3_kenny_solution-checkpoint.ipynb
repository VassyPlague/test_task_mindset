```python
def horizontal_crop(img, original_image):
    h = 0
    w = int(img.shape[1]/4)
    high = True
    counter = 0
    
    high_h = 0
    low_h = 0
    while h < img.shape[0]:
        while w < int (img.shape[1] - img.shape[1]/6):
            if img[h][w] == 255 and high == True:
                high_h = h
                high = False
                w = int(img.shape[1]/4)
                break
            if img[h][w] == 255:
                counter += 1
                w = int(img.shape[1]/4)
                break
            w += 1
        w = int(img.shape[1]/4)
        h += 1
    print (high_h)
    plt.imshow(img[high_h: high_h + counter])
    #plt.imshow(img)
    return original_image[high_h: high_h + counter]
```

```python
h_cropped = horizontal_crop(bordered_canny_img, g_b)
```

```python
def vert_crop(img):
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ret, edge = cv2.threshold(img_gray, 180, 255, cv2.THRESH_BINARY)
    w = int(edge.shape[1]/6)
    h = 0
    counter = 0
    first = 0 
    #plt.imshow(edge)
    #input ()
    while w < edge.shape[1] - int(edge.shape[1]/4):
        while h < edge.shape[0] - int(edge.shape[0]/2):
            if edge[h][w] == 0:
                if first == 0:
                    first = w
                counter += 1
                h += 1
                break
            h += 1
        h = 0
        w += 1
        #
    plt.imshow(img[0:edge.shape[0], first : first + counter + int(edge.shape[1]/50) ])
    a = img[0:edge.shape[0], first : first + counter + int(edge.shape[1]/50)]
    return a
    # реализовать метод с двумя полосками
```

```python
string_name = vert_crop(resized_img)
```

```python
from skimage.transform import resize
def char_cropper(img):
    output = []
    w = 0
    while w < img.shape[1]:
        try:
            output.append(resize(img[0:img.shape[0], w : w + 9], (32,32)))
        except ValueError:
            pass
        w += 9
        
    return output
```

```python
chars = char_cropper(string_name)
```

```python
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray
```
